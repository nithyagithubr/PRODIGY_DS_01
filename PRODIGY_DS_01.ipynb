```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("population_dataset.csv", skiprows=4)

df.head()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns[-10:]
```

```python
plt.figure()

plt.hist(df['2020'].dropna(), bins=10)

plt.title("Population Distribution (2020)")
plt.xlabel("Population")
plt.ylabel("Frequency")

plt.show()
```

```python
top10 = df[['Country Name', '2020']].dropna()

top10 = top10.sort_values(by='2020', ascending=False).head(10)

plt.figure()

plt.bar(top10['Country Name'], top10['2020'])

plt.xticks(rotation=45)
plt.title("Top 10 Countries by Population (2020)")
plt.xlabel("Country")
plt.ylabel("Population")

plt.show()
```

```python
plt.figure()

sns.barplot(x='Country Name', y='2020', data=top10)

plt.xticks(rotation=45)
plt.title("Top 10 Countries (Seaborn)")

plt.show()
```
